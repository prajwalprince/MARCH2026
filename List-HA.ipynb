# Exercise Set: Python List Methods
# Task 1: Add Your Favorite Food

```python
foods = ["pizza", "burger", "pasta"]
foods.append("taco")
foods.append("biriyani")
print(foods)
```

# Task 2: Add Multiple Items
# Add "green" and "yellow" at once.

```python
colors = ["red", "blue"]
colors.extend(["green", "yellow"]   )
print(colors)
```

# Task 3: Insert at Specific Position


# Insert 3 in the correct position.

```python
numbers = [1, 2, 4, 5]
numbers.insert(2, 3)
print(numbers)
```

# Task 4: Remove an Item
# Remove "mango".

```python
fruits = ["apple", "banana", "mango", "grapes"]
fruits.remove("banana")
print(fruits)
```

# Task 5: Remove Last Item

```python
items = ["pen", "book", "eraser"]
items.pop()
print(items)
```

Task 6: Find Index
# Find the position of "lion".

```python
animals = ["cat", "dog", "lion", "tiger"]
animals.index("lion")
print(animals.index("lion"))    
```

# Task 7: Count Occurrences
# Count how many times 2 appears.

```python
nums = [1, 2, 2, 3, 2, 4]
print(nums.count(2))
```

# Task 8: Sort the List(Ascending)

```python
marks = [85, 45, 90, 60]
marks.sort()
print(marks)
```

# Task 9: Reverse the List

```python
letters = ["a", "b", "c", "d"]
letters.reverse()
print(letters)
```

# Task 10: Copy the List

```python
original = [10, 20, 30]
print(original)
copy_cat = original.copy()
print(copy_cat)
```

Task 11: Clear the List

```python
data = [1, 2, 3, 4]
data.clear()
print(data)
```

Task 12: Extend a List

```python
list1 = [1, 2]
list2 = [3, 4]
list1.extend(list2)
print(list1)
```
