# len() And Indexing


```python
Course="AI And Data Science"
print("Length of variable is : ",len(Course))
x=0
y=-(len(Course))
for i in range(len(Course)):
  print(x," -> ",Course[i]," <- ",y,"\n")
  x+=1
  y+=1
```

# String Functions


```python
# upper()
Course="AI And Data Science"
print(Course.upper())

# lower()
my_movie="  *DETACHMENT (2011) PSYCHOLOGICAL DRAMA BY TONY KAYE* "
print(my_movie.lower())

# title()
print(Course.title())

# Strip()
print(my_movie.strip())

# lstrip()
print(my_movie.lstrip())

# rstrip()
print(my_movie.rstrip())

# split(sep,maxsplit)
breaking=my_movie.split()
print(breaking)

breaking1=my_movie.split(" ")
print(breaking1)

breaking2=my_movie.split(" ",2)
print(breaking2)
# join
print(" ".join(breaking))

print(" ".join(breaking1))

print(" ".join(breaking2))

# Slice
print(my_movie[0:10])

print(my_movie[0:10:2])

print(my_movie[::-1])
```
