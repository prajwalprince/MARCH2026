```python
# Question 6: Check if 97 is a prime number
print("\nQuestion 6: Check if 97 is a prime number")
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
print(f"Is 97 prime? {is_prime(97)}")
```
