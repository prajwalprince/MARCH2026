```python
# Question 41: Check if string contains balanced parentheses: "((()))"
print("\nQuestion 41: Check if string contains balanced parentheses: '((()))'")
# Your code here
text = "((()))"
count = 0
is_balanced = True
for char in text:
    if char == '(':
        count += 1
    elif char == ')':
        count -= 1
    if count < 0:
        is_balanced = False
        break
is_balanced = is_balanced and count == 0
print(f"Are parentheses balanced? {is_balanced}")
```

```python
# Question 43: Find the longest common substring between "programming" and "grammar"
print("\nQuestion 43: Find the longest common substring between 'programming' and 'grammar'")
# Your code here
str1 = "programming"
str2 = "grammar"
longest = ""
for i in range(len(str1)):
    for j in range(len(str1), i, -1):
        substr = str1[i:j]
        if substr in str2 and len(substr) > len(longest):
            longest = substr
print(f"Longest common substring: '{longest}'")
```

```python
# Question 46: Convert "hello world" to Pig Latin
print("\nQuestion 46: Convert 'hello world' to Pig Latin")
# Your code here
def to_pig_latin(word):
    vowels = 'aeiouAEIOU'
    if word[0] in vowels:
        return word + 'way'
    else:
        for i, char in enumerate(word):
            if char in vowels:
                return word[i:] + word[:i] + 'ay'
        return word + 'ay'

text = "hello world"
words = text.split()
pig_latin = ' '.join(to_pig_latin(word) for word in words)
print(f"Pig Latin: {pig_latin}")
```

```python
# Question 47: Check if string is a valid IPv4 address: "192.168.1.1"
print("\nQuestion 47: Check if string is a valid IPv4 address: '192.168.1.1'")
# Your code here
import re
ip = "192.168.1.1"
pattern = r'^(\d{1,3}\.){3}\d{1,3}$'
if re.match(pattern, ip):
    parts = ip.split('.')
    is_valid = all(0 <= int(part) <= 255 for part in parts)
else:
    is_valid = False
print(f"Is '{ip}' a valid IPv4 address? {is_valid}")
```

```python
# Question 50: Check if string is a valid credit card number: "4532015112830366"
print("\nQuestion 50: Check if string is a valid credit card number: '4532015112830366'")
# Your code here
def luhn_algorithm(card_number):
    digits = [int(d) for d in card_number if d.isdigit()]
    doubled = []
    for i, digit in enumerate(reversed(digits)):
        if i % 2 == 1:
            digit *= 2
            if digit > 9:
                digit -= 9
        doubled.append(digit)
    return sum(doubled) % 10 == 0

card = "4532015112830366"
is_valid = luhn_algorithm(card)
print(f"Is '{card}' a valid credit card number? {is_valid}")
```
